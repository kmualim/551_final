```python
from tensorflow.contrib.layers.python.layers import initializers
import tensorflow.contrib.slim as slim

slim = tf.contrib.slim
epsilon = 1e-9
log_dir = './log/train'
batch_size = 24
seed = 12345
```

```python
# use slim to create CNN layer
def relu_conv1(inputs, kernel, out_channels, stride, padding, name, is_train=True, activation_fn=None):
    tf.reset_default_graph()
    with slim.arg_scope([slim.conv2d], trainable=is_train):
        with tf.variable_scope(name) as scope:
            output = slim.conv2d(inputs,
                           num_outputs=out_channels,
                           kernel_size=kernel, stride=stride, padding=padding,
                           scope=scope, activation_fn=activation_fn)
            tf.logging.info(f"{name} output shape: {output.get_shape()}")
    return output
```

```python
# construct the primary capsule layer using the standard CNN layer
def primary_capsules(inputs, kernel_size, out_capsules, stride, padding, pose_shape, name):
    with tf.variable_scope(name) as scope:
        
        # create pose
        poses = relu_conv1(
            inputs,
            kernel_size,
            out_capsules * pose_shape[0] * pose_shape[1],
            stride,
            padding=padding,
            name='all_poses'
        )

        input_shape = inputs.get_shape()

        # Create 4x4 pose matrix
        poses = tf.reshape(
            poses,
            shape=[-1, input_shape[-3], input_shape[-2], out_capsules, pose_shape[0], pose_shape[1]],
            name='poses'
        )

        # Create activation for input capsules
        activations = relu_conv1(
            inputs,
            kernel_size,
            out_capsules,
            stride,
            padding=padding,
            activation_fn=tf.sigmoid,
            name='activation'
        )

        tf.summary.histogram(
            'activations', activations
        )

    return poses, activations
```

```python
# creates a primary_capsule layer from a standard conv layer
# use tiling and convolution to make pose matrices and activations the righ tsize for routing
def kernel_tile(inputs, kernel, stride):   
    input_shape = inputs.get_shape()
    size = input_shape[4]*input_shape[5] if len(input_shape)>5 else 1
    inputs = tf.reshape(inputs, shape=[-1, input_shape[1], input_shape[2], input_shape[3]*size])


    input_shape = inputs.get_shape()
    tile_filter = np.zeros(shape=[kernel, kernel, input_shape[3],
                                  kernel * kernel], dtype=np.float32)
    for i in range(kernel):
        for j in range(kernel):
            tile_filter[i, j, :, i * kernel + j] = 1.0 

    tile_filter_op = tf.constant(tile_filter, dtype=tf.float32)

    output = tf.nn.depthwise_conv2d(inputs, tile_filter_op, strides=[
                                    1, stride, stride, 1], padding='VALID')

    output_shape = output.get_shape()
    output = tf.reshape(output, shape=[-1, output_shape[1], output_shape[2], input_shape[3], kernel * kernel])
    output = tf.transpose(output, perm=[0, 1, 2, 4, 3])

    return output
```

```python
def load_batch(images, labels, batch_size=32):
    batch_img, batch_labels = [], []
    
    for img in images:
        batch_img.append(img)
        
    for lab in labels:
        batch_labels.append(lab)
    
    return batch_img, batch_labels


def run(img, lab, seed, batch_size, log_dir):
    tf.logging.set_verbosity(tf.logging.INFO)
    np.random.seed(seed)
    tf.set_random_seed(seed)

    iterations_per_epoch = len(dataset) // batch_size

    images, labels = load_batch(img, lab, batch_size)

    one_hot_labels = slim.one_hot_encoding(labels, 5)

    poses, activations = capsules_net(images, num_classes=5, iterations=3, batch_size=batch_size, name='capsules_em')

    global_step = tf.train.get_or_create_global_step()
    loss = spread_loss(
        one_hot_labels, activations, iterations_per_epoch, global_step, name='spread_loss'
    )
    tf.summary.scalar('losses/spread_loss', loss)

    optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
    train_tensor = slim.learning.create_train_op(
        loss, optimizer, global_step=global_step, clip_gradient_norm=4.0
    )

    slim.learning.train(
        train_tensor,
        logdir=log_dir,
        log_every_n_steps=10,
        save_summaries_secs=60,
        saver=tf.train.Saver(max_to_keep=2),
        save_interval_secs=600,
    )
```

```python
# compute vote V = M * W
def mat_transform(inputs, output_cap_size, size):
    caps_num_i = int(inputs.get_shape()[1]) 
    output = tf.reshape(inputs, shape=[size, caps_num_i, 1, 4, 4]) 

    w = slim.variable(
        'w',
        shape=[1, caps_num_i, output_cap_size, 4, 4],
        dtype=tf.float32,
        initializer=tf.truncated_normal_initializer(mean=0.0, stddev=1.0)
    ) 
    
    w = tf.tile(w, [size, 1, 1, 1, 1])  

    output = tf.tile(output, [1, 1, output_cap_size, 1, 1]) 
    votes = tf.matmul(output, w) 
    votes = tf.reshape(votes, [size, caps_num_i, output_cap_size, 16])

    return votes
```

```python
# coordinate addition
def coord_addition(votes, H, W):
    coordinate_offset_hh = tf.reshape(
      (tf.range(H, dtype=tf.float32) + 0.50) / H, [1, H, 1, 1, 1]
    )
    coordinate_offset_h0 = tf.constant(
      0.0, shape=[1, H, 1, 1, 1], dtype=tf.float32
    )
    coordinate_offset_h = tf.stack(
      [coordinate_offset_hh, coordinate_offset_h0] + [coordinate_offset_h0 for _ in range(14)], axis=-1
    )  

    coordinate_offset_ww = tf.reshape(
      (tf.range(W, dtype=tf.float32) + 0.50) / W, [1, 1, W, 1, 1]
    )
    coordinate_offset_w0 = tf.constant(
      0.0, shape=[1, 1, W, 1, 1], dtype=tf.float32
    )
    coordinate_offset_w = tf.stack(
      [coordinate_offset_w0, coordinate_offset_ww] + [coordinate_offset_w0 for _ in range(14)], axis=-1
    ) 

    votes = votes + coordinate_offset_h + coordinate_offset_w

    return votes
```

```python

# creates convolution layer from a primary_capsule or a relu_conv1
# convolute pose matrices
# compute votes
# run em routing
def conv_capsule(inputs, shape, strides, iterations, batch_size, name):
    inputs_poses, inputs_activations = inputs
    
    with tf.variable_scope(name) as scope:
        stride = strides[1] 
        i_size = shape[-2]  
        o_size = shape[-1]  
        pose_size = inputs_poses.get_shape()[-1]  

        # fit the input pose matrices to the dimension of out caps
        # used for multiplication with transformation matrix for votes
        inputs_poses = kernel_tile(inputs_poses, 3, stride) 

        # fit activations for em routing
        inputs_activations = kernel_tile(inputs_activations, 3, stride)
        spatial_size = int(inputs_activations.get_shape()[1]) 

        inputs_poses = tf.reshape(inputs_poses, shape=[-1, 3 * 3 * i_size, 16])  
        inputs_activations = tf.reshape(inputs_activations, shape=[-1, spatial_size, spatial_size, 3 * 3 * i_size])

        with tf.variable_scope('votes') as scope:
          
            # create votes : V = M * W
            votes = mat_transform(inputs_poses, o_size, size=batch_size*spatial_size*spatial_size)
        
            votes_shape = votes.get_shape()
            votes = tf.reshape(votes, shape=[batch_size, spatial_size, spatial_size, votes_shape[-3], votes_shape[-2], votes_shape[-1]])
            tf.logging.info(f"{name} votes shape: {votes.get_shape()}")

        with tf.variable_scope('routing') as scope:
            #  Each output capsule has beta_v, beta_a
            beta_v = tf.get_variable(
                name='beta_v',
                shape=[1, 1, 1, o_size],
                dtype=tf.float32,
                initializer=initializers.xavier_initializer()
              )
            
            beta_a = tf.get_variable(
              name='beta_a', 
              shape=[1, 1, 1, o_size],
              dtype=tf.float32,
              initializer=initializers.xavier_initializer()
          )

            # create pose and activation via em routing
            poses, activations = matrix_capsules_em_routing(
                votes,
                inputs_activations,
                beta_v,
                beta_a,
                iterations,
                name='em_routing'
          )

            # make 4 x 4 
            poses_shape = poses.get_shape()
            poses = tf.reshape(
                poses,
                [poses_shape[0], poses_shape[1], poses_shape[2], poses_shape[3], pose_size, pose_size]
              )

        tf.logging.info(f"{name} pose shape: {poses.get_shape()}")
        tf.logging.info(f"{name} activations shape: {activations.get_shape()}")

        return poses, activations
```

```python
def class_capsules(inputs, num_classes, iterations, batch_size, name):
    inputs_poses, inputs_activations = inputs 

    inputs_shape = inputs_poses.get_shape()
    spatial_size = int(inputs_shape[1])  
    pose_size = int(inputs_shape[-1])    
    i_size = int(inputs_shape[3])        

    inputs_poses = tf.reshape(inputs_poses, shape=[batch_size*spatial_size*spatial_size, inputs_shape[-3], inputs_shape[-2]*inputs_shape[-2] ])

    with tf.variable_scope(name) as scope:
        with tf.variable_scope('votes') as scope:
            # get votes
            votes = mat_transform(inputs_poses, num_classes, size=batch_size*spatial_size*spatial_size)
            tf.logging.info(f"{name} votes shape: {votes.get_shape()}")

            votes = tf.reshape(votes, shape=[batch_size, spatial_size, spatial_size, i_size, num_classes, pose_size*pose_size])

            votes = coord_addition(votes, spatial_size, spatial_size)

            tf.logging.info(f"{name} votes shape with coord addition: {votes.get_shape()}")

        with tf.variable_scope('routing') as scope:
            # beta_v and beta_a for each out capsule
            beta_v = tf.get_variable(
                name='beta_v', shape=[1, num_classes], dtype=tf.float32,
                initializer=initializers.xavier_initializer()
            )
            beta_a = tf.get_variable(
                name='beta_a', shape=[1, num_classes], dtype=tf.float32,
                initializer=initializers.xavier_initializer()
            )

            # votes 
            votes_shape = votes.get_shape()
            votes = tf.reshape(votes, shape=[batch_size, votes_shape[1] * votes_shape[2] * votes_shape[3], votes_shape[4], votes_shape[5]] )

            
            inputs_activations = tf.reshape(inputs_activations, shape=[batch_size,
                                                                       votes_shape[1] * votes_shape[2] * votes_shape[3]])

            poses, activations = matrix_capsules_em_routing(
                votes, inputs_activations, beta_v, beta_a, iterations, name='em_routing'
            )

        
        poses = tf.reshape(poses, shape=[batch_size, num_classes, pose_size, pose_size] )

        return poses, activations
```

```python
# get pose and activation of outputs via em routing for cap i to cap j
def matrix_capsules_em_routing(votes, i_activations, beta_v, beta_a, iterations, name):
    
    votes_shape = votes.get_shape().as_list()
    with tf.variable_scope(name) as scope:
        rr = tf.constant(
            1.0/votes_shape[-2],
            shape=votes_shape[-3:-1] + [1],
            dtype=tf.float32
        )
        
        # expand dimensions
        i_activations = i_activations[..., tf.newaxis, tf.newaxis]

        # expand dimensions
        beta_v = beta_v[..., tf.newaxis, :, tf.newaxis]
        beta_a = beta_a[..., tf.newaxis, :, tf.newaxis]
        
        # M-Step from cap i to cap j
        def m_step(rr, votes, i_activations, beta_v, beta_a, inverse_temperature):
            fd

            rr_prime = rr * i_activations

            # sum input capsules i
            rr_prime_sum = tf.reduce_sum(rr_prime, axis=-3, keep_dims=True, name='rr_prime_sum')

            # mean of out capsules
            o_mean = tf.reduce_sum(
                rr_prime * votes,
                axis=-3,
                keep_dims=True
              ) / rr_prime_sum

            # compute standard deviation
            o_stdv = tf.sqrt(
                tf.reduce_sum(
                    rr_prime * tf.square(votes - o_mean), axis=-3, keep_dims=True
                ) / rr_prime_sum
            )

            o_cost_h = (beta_v + tf.log(o_stdv + epsilon)) * rr_prime_sum

            o_cost = tf.reduce_sum(o_cost_h, axis=-1, keep_dims=True)
            o_cost_mean = tf.reduce_mean(o_cost, axis=-2, keep_dims=True)
            o_cost_stdv = tf.sqrt(
                    tf.reduce_sum(
                          tf.square(o_cost - o_cost_mean), axis=-2, keep_dims=True
                    ) / o_cost.get_shape().as_list()[-2]
            )
            o_activations_cost = beta_a + (o_cost_mean - o_cost) / (o_cost_stdv + epsilon)


            o_activations = tf.sigmoid(
                inverse_temperature * o_activations_cost
            )

            return o_mean, o_stdv, o_activations
        
        # E-Step
        def e_step(o_mean, o_stdv, o_activations, votes):
            o_p_unit0 = - tf.reduce_sum(
                tf.square(votes - o_mean) / (2 * tf.square(o_stdv)), axis=-1, keep_dims=True
            )

            o_p_unit2 = - tf.reduce_sum(
                tf.log(o_stdv + epsilon), axis=-1, keep_dims=True
            )

            # prob of the Vh i to j
            o_p = o_p_unit0 + o_p_unit2

            zz = tf.log(o_activations + epsilon) + o_p
            rr = tf.nn.softmax(
                zz, dim=len(zz.get_shape().as_list())-2
            )

            return rr
        

        # inverse temperature
        it_min = 1.0
        it_max = min(iterations, 3.0)
        for it in range(iterations):
            inverse_temperature = it_min + (it_max - it_min) * it / max(1.0, iterations - 1.0)
            o_mean, o_stdv, o_activations = m_step(
                rr, votes, i_activations, beta_v, beta_a, inverse_temperature=inverse_temperature
            )
            if it < iterations - 1:
                rr = e_step(
                  o_mean, o_stdv, o_activations, votes
                )

        poses = tf.squeeze(o_mean, axis=-3)
        activations = tf.squeeze(o_activations, axis=[-3, -1])
    return poses, activations
```

```python

# creates convolution layer from a primary_capsule or a relu_conv1
# convolute pose matrices
# compute votes
# run em routing
def conv_capsule(inputs, shape, strides, iterations, batch_size, name):
    inputs_poses, inputs_activations = inputs
    
    with tf.variable_scope(name) as scope:
        stride = strides[1] 
        i_size = shape[-2]  
        o_size = shape[-1]  
        pose_size = inputs_poses.get_shape()[-1]  

        # fit the input pose matrices to the dimension of out caps
        # used for multiplication with transformation matrix for votes
        inputs_poses = kernel_tile(inputs_poses, 3, stride) 

        # fit activations for em routing
        inputs_activations = kernel_tile(inputs_activations, 3, stride)
        spatial_size = int(inputs_activations.get_shape()[1]) 

        inputs_poses = tf.reshape(inputs_poses, shape=[-1, 3 * 3 * i_size, 16])  
        inputs_activations = tf.reshape(inputs_activations, shape=[-1, spatial_size, spatial_size, 3 * 3 * i_size])

        with tf.variable_scope('votes') as scope:
          
            # create votes : V = M * W
            votes = mat_transform(inputs_poses, o_size, size=batch_size*spatial_size*spatial_size)
        
            votes_shape = votes.get_shape()
            votes = tf.reshape(votes, shape=[batch_size, spatial_size, spatial_size, votes_shape[-3], votes_shape[-2], votes_shape[-1]])
            tf.logging.info(f"{name} votes shape: {votes.get_shape()}")

        with tf.variable_scope('routing') as scope:
            #  Each output capsule has beta_v, beta_a
            beta_v = tf.get_variable(
                name='beta_v',
                shape=[1, 1, 1, o_size],
                dtype=tf.float32,
                initializer=initializers.xavier_initializer()
              )
            
            beta_a = tf.get_variable(
              name='beta_a', 
              shape=[1, 1, 1, o_size],
              dtype=tf.float32,
              initializer=initializers.xavier_initializer()
          )

            # create pose and activation via em routing
            poses, activations = matrix_capsules_em_routing(
                votes,
                inputs_activations,
                beta_v,
                beta_a,
                iterations,
                name='em_routing'
          )

            # make 4 x 4 
            poses_shape = poses.get_shape()
            poses = tf.reshape(
                poses,
                [poses_shape[0], poses_shape[1], poses_shape[2], poses_shape[3], pose_size, pose_size]
              )

        tf.logging.info(f"{name} pose shape: {poses.get_shape()}")
        tf.logging.info(f"{name} activations shape: {activations.get_shape()}")

        return poses, activations
```

```python
# create capsule architecture
def capsules_net(inputs, num_classes, iterations, batch_size, name='em_caps'):
    with tf.variable_scope(name) as scope:
        
        # ReLU layer 1
        # conv 5x5 filters, 32 output channels, 2 strides with padding, ReLU
        nets = relu_conv1(
            inputs,
            kernel=5,
            out_channels=32,
            stride=2,
            padding='SAME',
            activation_fn=tf.nn.relu,
            name='relu_conv1'
        )

        # primary capsules
        # capsule 1x1 filter, 32 output capsules, 1 stride without padding
        nets = primary_capsules(
            nets,
            kernel_size=1,
            out_capsules=32,
            stride=1,
            padding='VALID',
            pose_shape=[4, 4],
            name='primary_capsules'
        )

        # conv capsules 1
        # conv capsule, 3x3 kernels, 2 strides with padding
        nets = conv_capsule(
            nets,
            shape=[3, 3, 32, 32],
            strides=[1, 2, 2, 1],
            iterations=iterations,
            batch_size=batch_size,
            name='conv_caps1'
        )

        # conv capsules 2
        # conv capsule, 3x3 kernels, 1 string without padding
        nets = conv_capsule(
            nets,
            shape=[3, 3, 32, 32],
            strides=[1, 1, 1, 1],
            iterations=iterations,
            batch_size=batch_size,
            name='conv_caps2'
        )

        # final layer
        # 1x1 conv, 10 outpus 
        nets = class_capsules(
            nets,
            num_classes,
            iterations=iterations,
            batch_size=batch_size,
            name='class_capsules'
        )

        poses, activations = nets

    return poses, activations

# spread loss
def spread_loss(labels, activations, iterations_per_epoch, global_step, name):
    margin = tf.train.piecewise_constant(
        tf.cast(global_step, dtype=tf.int32),
        boundaries = [
            (iterations_per_epoch * x) for x in range(1, 8)],
        values = [x / 10.0 for x in range(2, 10)]
    )

    activations_shape = activations.get_shape().as_list()

    with tf.variable_scope(name) as scope:
        # labels for masks
        mask_t = tf.equal(labels, 1)      
        mask_i = tf.equal(labels, 0)      

        # activation for label
        activations_t = tf.reshape(
            tf.boolean_mask(activations, mask_t),
            shape=(tf.shape(activations)[0], 1)
        )

        # other activations
        activations_i = tf.reshape(
            tf.boolean_mask(activations, mask_i),
            [tf.shape(activations)[0], activations_shape[1] - 1]
        )

        l = tf.reduce_sum(
            tf.square(
                tf.maximum(
                    0.0,
                    margin - (activations_t - activations_i)
                )
            )
        )
        tf.losses.add_loss(l)

        return l
```

```python
def load_batch(images, labels, batch_size=32):
    batch_img, batch_labels = [], []
    
    for img in images:
        batch_img.append(img)
        
    for lab in labels:
        batch_labels.append(lab)
    
    return batch_img, batch_labels


def main(_):
    tf.logging.set_verbosity(tf.logging.INFO)
    np.random.seed(seed)
    tf.set_random_seed(seed)

    iterations_per_epoch = len(train) // batch_size

    images, labs = load_batch(train, labels, batch_size)

    one_hot_labels = slim.one_hot_encoding(labs, 5)

    poses, activations = capsules_net(images, num_classes=5, iterations=3, batch_size=batch_size, name='capsules_em')

    global_step = tf.train.get_or_create_global_step()
    loss = spread_loss(
        one_hot_labels, activations, iterations_per_epoch, global_step, name='spread_loss'
    )
    tf.summary.scalar('losses/spread_loss', loss)

    optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
    train_tensor = slim.learning.create_train_op(
        loss, optimizer, global_step=global_step, clip_gradient_norm=4.0
    )

    slim.learning.train(
        train_tensor,
        logdir=log_dir,
        log_every_n_steps=10,
        save_summaries_secs=60,
        saver=tf.train.Saver(max_to_keep=2),
        save_interval_secs=600,
    )
```
